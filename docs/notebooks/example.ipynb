```python
%load_ext autoreload
%autoreload 2
from multi_condition_comparisions.tl.de import StatsmodelsDE
import scanpy as sc
```

```python
adata = sc.datasets.pbmc68k_reduced()
adata.obs
```

```python
mod = StatsmodelsDE(adata, "~ phase")
```

```python
mod.fit()
```

```python
mod.test_contrasts({"G2M_vs_S": [0, -1, 1]})
```

```python
mod.test_contrasts({"G2M_vs_S": mod.cond(phase="S") - mod.cond(phase="G2M")})
```

```python
mod.test_contrasts({"G2M_vs_S": mod.contrast("phase", "S", "G2M")})
```
